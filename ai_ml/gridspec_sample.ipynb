```python
# ライブラリの読み込み
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.gridspec import  GridSpec
from sklearn.metrics import confusion_matrix
```

```python
# 描画用の点を乱数で発生させる
np.random.seed(329)
x = np.random.randn(1000)
y = np.random.randn(1000)
```

```python
# GridSpecで描画領域を確保
fig = plt.figure(figsize=(7,5))
gs = GridSpec(2, 2, figure=fig, width_ratios=(6, 1), height_ratios=(1, 4))

# 左下の散布図
ax0 = plt.subplot(gs[1, 0])
ax0.scatter(x, y, color="c", alpha=0.2)
ax0.grid()
#ax0.xaxis.tick_top()

# 右上のヒストグラム
ax01 = plt.subplot(gs[1, 1])
ax01.hist(y, 40, orientation="horizontal")
#ax01.invert_xaxis()
ax01.yaxis.tick_right()

# 左上のヒストグラム
ax10 = plt.subplot(gs[0, 0])
ax10.hist(x, 60)
#ax10.invert_yaxis()
ax10.xaxis.tick_top()

plt.tight_layout()
plt.show()
```

回帰(数値の予測)の時は上記の様な散布図で可視化すると結果が一目瞭然なのではんよ。
分類(カテゴリカルデータの予測)の時は混合行列(confusion_matrix)をヒートマップにするのが良いでしょう。

# セルを結合する
https://matplotlib.org/gallery/subplots_axes_and_figures/gridspec_multicolumn.html#sphx-glr-gallery-subplots-axes-and-figures-gridspec-multicolumn-py

```python
def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)

fig = plt.figure(constrained_layout=True)

gs = GridSpec(3, 3, figure=fig)
ax1 = fig.add_subplot(gs[0, :])
# identical to ax1 = plt.subplot(gs.new_subplotspec((0, 0), colspan=3))
ax2 = fig.add_subplot(gs[1, :-1])
ax3 = fig.add_subplot(gs[1:, -1])
ax4 = fig.add_subplot(gs[-1, 0])
ax5 = fig.add_subplot(gs[-1, -2])

fig.suptitle("GridSpec")
format_axes(fig)

plt.show()
```
